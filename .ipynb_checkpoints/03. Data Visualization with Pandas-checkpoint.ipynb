```python
# anscombe 데이터셋을 시각화를 통해 matplotlib 시각화 방법과 함께 시각화 중요성 알아보기
import seaborn as sns

# 샘플 데이터셋 목록 확인
sns.get_dataset_names()
```

```python
# anscombe 데이터셋 로드
anscombe = sns.load_dataset("anscombe")
print(type(anscombe))
anscombe
```

```python
# 데이터셋 구분자 확인
anscombe['dataset'].unique()
```

```python
# 데이터셋의 개수 확인
anscombe['dataset'].nunique()
```

```python
# 하위 데이터셋 구분
# anscombe['dataset']=='I'
ds_1 = anscombe[anscombe['dataset']=='I']
ds_2 = anscombe[anscombe['dataset']=='II']
ds_3 = anscombe[anscombe['dataset']=='III']
ds_4 = anscombe[anscombe['dataset']=='IV']
```

```python
# 데이터 셋 I
ds_1
```

```python
# 데이터셋 I의 기초 통계량
ds_1.describe()
```

```python
# 데이터셋 II의 기초 통계량
ds_2.describe()
```

```python
# 시각화 : dataset I 시각화
import matplotlib.pyplot as plt
plt.plot(ds_1['x'],ds_1['y'],'o')
```

```python
import matplotlib.pyplot as plt
# 4분할 그래프
# 그래프의 기본 틀 생성: figure
fig = plt.figure()

# 기본 틀에 격자 추가
axes1 = fig.add_subplot(2, 2, 1)           # 2행 2열의 그래프 중 1번째 영역
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)
```

```python
# 격자에 그래프 그리기
axes1.plot(ds_1['x'], ds_1['y'], 'o')
axes2.plot(ds_2['x'], ds_2['y'], '*', color="orange")
axes3.plot(ds_3['x'], ds_3['y'], 'v', color="green")
axes4.plot(ds_4['x'], ds_4['y'], '^', color="red")

fig
```

```python
# 서브 플롯들에 타이틀 추가
axes1.set_title("dataset I")
axes2.set_title("dataset II")
axes3.set_title("dataset III")
fig
```

```python
# 전체틀(figure)의 타이틀 추가
fig.suptitle("Anscombe Dataset")
fig
```

```python
# 산점도 그래프(Scatter Plot)
# mtcars 데이터셋
import pandas as pd
mtcars = pd.read_csv("./data/mtcars.csv")
mtcars
```

```python
# EDA 방식의 데이터 분석
print("데이터의 크기:", mtcars.shape)          # 32행 12열
```

```python
# 앞쪽 데이터 확인
mtcars.head()
```

```python
# 뒤쪽 데이터 확인
mtcars.tail()
```

```python
# 기초 통계량
mtcars.describe()
```

```python
# mpg(y축) vs wt(x축) -> 산점도로 시각화
import matplotlib.pyplot as plt

fig = plt.figure()                             # 틀
axe = fig.add_subplot(1, 1, 1)                 # 1행1열의 1번째 축
axe.scatter(mtcars['wt'], mtcars['mpg'])
axe.set_title("mtcars wt vs mpg")
axe.set_xlabel("Weight")
axe.set_ylabel("Miles per Gallog")

plt.show()
```
